<div style="text-align: justify"> The prisoner's dilemma is a decision-analysis conundrum when two people acting in their own self-interests do not produce the optimal outcome.

In the traditional prisoner's dilemma, individuals find that betraying the group yields greater rewards than cooperating, and so they have an incentive to make decisions that have a negative impact on the group.

This game is an example of this Prisonerâ€™s Dilemma based of an environmental conservation issue. Two rival companies, based on different banks of the same river, pour their water waste to said river. Both companies are studying the possibility to invest in pollution control equipment that would measure the levels of toxicity of the waste, compare those to the levels in the river and report the damage that it could potentially do to the ecosystem. 

If both companies invest in the equipment, the resource will be protected and both companies will benefit.
If only one of the companies invests in the equipment, they will report the other company for over polluting the river and take out the competition. 
If neither of the companies invests in the equipment, the river will be destroyed and the companies will have to relocate, which is a much higher cost than the equipment.
